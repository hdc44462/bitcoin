```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
coin2012 = pd.read_csv('C:\\Users\\dchua\\Desktop\\bitcoin201201.csv')
coin2014 = pd.read_csv('C:\\Users\\dchua\\Desktop\\bitcoin201404.csv')
```

```python
coin = coin2012.append(coin2014,sort = False)
coin.reset_index(inplace = True)
```

```python
coin.shape
```

```python
coin.isnull().sum()
```

```python
coin.Timestamp.dtypes
```

```python
coin['date'] = pd.to_datetime(coin['Timestamp'],unit='s').dt.date
```

```python
#daily average price for bitcoin
day_trans = coin.groupby('date').mean()
sns.set(rc={'figure.figsize':(15,8)})
ax = day_trans[['Weighted_Price','Volume_(Currency)']].plot(secondary_y = 'Volume_(Currency)',legend = True)
```

```python
coin['month'] = coin.date.apply(lambda x: x.month)
coin['year'] = coin.date.apply(lambda x: x.year)
month_trans = coin.groupby(['year','month']).mean()
ax2 = month_trans[['Weighted_Price','Volume_(Currency)']].plot(secondary_y = 'Volume_(Currency)')
```
