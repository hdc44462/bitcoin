```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
from statsmodels.tsa.stattools import adfuller
adfuller(day_trans2014.Weighted_Price)
```

```python
#As pvalue 0.378 > 0.05, we cant reject the unstability of this time series. 
#first degree diff
D1_data = day_trans2014.Weighted_Price.diff().dropna()
```

```python
plot_acf(D1_data)
adfuller(D1_data) 
#p-value significantly < 0.05, first diff is stable.
```

```python
#seasonal ARIMA model
import statsmodels.api as sm
from statsmodels.tsa.stattools import acf
from statsmodels.tsa.stattools import pacf
from statsmodels.tsa.seasonal import seasonal_decompose
decomposition = sm.tsa.seasonal_decompose(day_trans2014.Weighted_Price,freq = 1)
fig = plt.figure()
fig = decomposition.plot()
fig.set_size_inches(12, 6)
```

```python
coin2014.Timestamp = pd.to_datetime(coin2014.Timestamp, unit='s')

# Resampling to daily frequency
coin2014.index = coin2014.Timestamp
coin2014 = coin2014.resample('D').mean()

# Resampling to monthly frequency
coin2014_month = coin2014.resample('M').mean()

# Resampling to annual frequency
coin2014_year = coin2014.resample('A-DEC').mean()

# Resampling to quarterly frequency
coin2014_Q = coin2014.resample('Q-DEC').mean()
```

```python
#SARIMA of coin2014
plt.figure(figsize=[15,7])
sm.tsa.seasonal_decompose(coin2014_month.Weighted_Price).plot()
print("DickeyFuller test: p=%f" % sm.tsa.stattools.adfuller(coin2014_month.Weighted_Price)[1])
plt.show()
```

```python
D1_data =coin2014_month.Weighted_Price.diff().dropna()
sm.tsa.seasonal_decompose(D1_data).plot()
adfuller(D1_data)
```
